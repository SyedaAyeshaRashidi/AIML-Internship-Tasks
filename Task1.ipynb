# Task 1: Exploring and Visualizing the Iris Dataset
**Objective:** To load, inspect, and visualize the dataset to understand trends.

## 1. Loading the Dataset

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Loading the dataset with error handling to ensure stability
try:
    df = sns.load_dataset('iris')
    print("Dataset loaded successfully!")
    # Displaying first 5 rows for initial inspection
    display(df.head())
except Exception as e:
    print(f"Error loading dataset: {e}")
```

## 2. Inspection and Statistics

```python
# Summary statistics table
print("\nSummary Statistics:")
display(df.describe())
```

## 3. Data Visualization

```python
# --- Graph 1: Scatter Plot ---
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x='sepal_length', y='petal_length', hue='species')
plt.title('Scatter Plot: Sepal vs Petal')
plt.show()  

# --- Graph 2: Histogram ---
df.hist(figsize=(10,8), color='teal', edgecolor='black')
plt.suptitle('Histograms for all Features')
plt.show() 

# --- Graph 3: Box Plot ---
plt.figure(figsize=(8,5))
sns.boxplot(data=df, x='species', y='petal_width')
plt.title('Box Plot: Petal Width by Species')
plt.show() 
```

Final Insights:

Data Structure: The dataset contains 150 samples with 4 numerical features and 1 categorical target (Species).

Species Separation: The Scatter plot shows that 'Setosa' is highly distinct from 'Versicolor' and 'Virginica'.

Outliers: Box plots indicate that some features, like sepal width, have potential outliers in certain species.
